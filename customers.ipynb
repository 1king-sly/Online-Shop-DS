Creating A Relationship between Products and Order Items

```python
products_dataset.columns
```

```python
order_items_dataset.columns
```

```python
product_order_dataset = pd.merge(products_dataset[["product_id","product_name","category","price","supplier_id"]],order_items_dataset[["order_item_id","order_id","product_id","quantity","price_at_purchase"]],on="product_id")


print(product_order_dataset)
```

```python
product_order_dataset.sort_values(["product_id"])
```

```python
product_order_dataset.plot(x="category", y="price", kind="bar")

plt.ylim(ymin=0)
plt.xlim(xmin=0)


plt.show()
```

```python
category_price = product_order_dataset.drop_duplicates(subset=["product_id"]).groupby("category")["price"].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(category_price["category"], category_price["price"], color='skyblue')
plt.xlabel("Category")
plt.ylabel("Total Price")
plt.title("Total Price by Product Category")
plt.xticks(rotation=45)
plt.show()
```
